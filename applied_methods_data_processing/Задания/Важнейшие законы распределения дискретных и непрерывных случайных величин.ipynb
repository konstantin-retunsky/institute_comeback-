```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

# Важнейшие законы распределения дискретных и непрерывных случайных величин
## Законы распределения дискретных случайных величин
### Распределение Бернулли
Для создания случайной величины, имеющей распределение Бернулли с параметром $p$, используется класс `bernoulli(p)` из модуля `stats` библиотеки `scipy`.
### Биномиальное распределение
Для создания случайной величины, имеющей биномиальное распределение с параметрами $n$ и $p$, используется класс `binomial(n, p)` из модуля `stats` библиотеки `scipy`.
### Распределение Пуассона
Для создания случайной величины, имеющей имеющей распределение Пуассона с параметром $\lambda$, используется класс `poisson(lamda)` из модуля `stats` библиотеки `scipy`.
## Законы распределения непрерывных случайных величин
### Равномерное распределение
Для создания равномерно распределенной на отрезке $[a,b]$ случайной величины используется класс `uniform(a, b-a)` из модуля `stats` библиотеки `scipy`.

Обратите внимание, что указываемыми параметрами являются левая граница промежутка и его длина, а не левая и правая границы промежутка.
### Показательное (экпоненциальное) распределение
Для создания случайной величины, имеющей показательное распределение с параметром $\lambda$, используется класс `expon(scale = 1 / lambda)` из модуля `stats` библиотеки `scipy`.
### Нормальное распределение
Для создания нормально распределенной случайной величины с параметрами $m$ и $\sigma$ используется класс `norm(m,sigma)` из модуля `stats` библиотеки `scipy`.
Если параметры нормального распределения не указаны, то создается случайная величина со стандратным нормальным законом распределения.

## Методы для работы со случайными величинами
### Массив из значений случайной величины
Для вычисления $n$ значений случайной величины используется метод `rvs(n)` соответствующего класса.

```python
# пример: массив из 10 значений случайной величины с показательными распределением с параметром lamda = 0.8
lamda = 0.8

# создаем случайную величину expon_rv c показательным распределением
expon_rv = sts.expon(scale=1/lamda)

expon_rv.rvs(10)
```

### Функция распределения
Для вычисления значения функции распределения $F(x)$ в точке $x$ используется метод `cdf(x)` соответствующего класса.

```python
# пример: построение графика функции распределения равномерно распределенной случайной величины с параметрами a = 1 и b = 4
a = 1
b = 4

# создаем равномерно распределенную случайную величину с именем uniform_rv
uniform_rv = sts.uniform(a, b-a)

# формируем массив из 100 абсцисс точек в диапазоне от 0 до 5
# выбор длины диапазона и количества точек зависит от конкретного распределения
x = np.linspace(0, 5, 100)

# считаем F(x) для каждого x
cdf = uniform_rv.cdf(x) # метод cdf может принимать и массив x

#cтроим график функции распределения
plt.plot(x, cdf)

# добавляем названия для осей и графика
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title("Функция распределения равномерного закона")
```

Для построения графика функции распределения в случае дискретной случайной величины используйте метод `step` библиотеки `matplotlib`.

### Плотность распределения
Для вычисления значения функции плотности распредления вероятностей $f(x)$ в точке $x$ используется метод `pdf(x)` соответствующего класса.

```python
# пример: построение графика плотности распределения нормально распределено случайной величины с параметрами m = 3 и sigma = 0.5
m = 2
sigma = 0.5

# создаем нормально распределенную случайную величину с именем norm_rv
norm_rv = sts.norm(m, sigma)

# формируем массив из 100 абсцисс точек в диапазоне от 0 до 4
# выбор длины диапазона и количества точек зависит от конкретного распределения
x = np.linspace(0, 4, 100)

# считаем fx) для каждого x
pdf = norm_rv.pdf(x) # метод pdf может принимать и массив x

#cтроим график плотности распределения
plt.plot(x, pdf)

# добавляем названия для осей и графика
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("Плотность распределения нормального закона")
```

### Функция вероятностей
Для дискретных случайных величин используется метод `pmf`, считающий вероятность $P(x=x)$ случайной величине $X$ принять значение $x$.

```python
# пример: построение многоугольника распределения случайной величины с биномиальным распределением параметрами n = 20 и p = 0.7
n = 20
p = 0.7
 
# создаем случайную величину binomial_rv с биномиальным распределением
binomial_rv = sts.binom(n, p)

# формируем массив x = 0, 1, 2, ..., 19, 20
# выбор длины диапазона и количества точек зависит от конкретного распределения
x = np.linspace(0, 20, 21)

# считаем P(X=x) для каждого x
pmf = binomial_rv.pmf(x) # метод pmf может принимать и массив x

#cтроим многоугольник распределения
plt.plot(x, pmf, 'o')

# добавляем названия для осей и графика
plt.ylabel('$P$')
plt.xlabel('$x$')
plt.title("Многоугольник распределения\nбиномиального закона")
```

### Основные числовые характеристики
Для вычисления основных числовых характеристик используется метод `stats()` соответствующего класса. Метод возвращает кортеж из значений математического ожидания и дисперсии.


Отдельно:
* для вычисления значения математического ожидания используется метод `mean()` соответствующего класса


* для вычисления значения дисперсии используется метод `var()` соответствующего класса 


* для вычисления значения среднего квадратического отклонения используется метод `std()` соответствующего класса


* для вычисления квантиля $x_p$  используется метод `ppf(p)` соответствующего класса (метод обратен методу `cdf(x)`). Для вычисления медианы (второго квартиля) можно отдельно использовать метод `median()` соответствующего класса 


```python
# пример: вычисление математического ожидания, дисперсии и среднего квадратического отклонения
# для ранее созданной случайной величины с равномерным распределением

mean_u, var_u = uniform_rv.stats()

print(f'M(X) = {mean_u}')
print(f'D(X) ≈ {var_u}')
print(f'𝜎(X) ≈ {uniform_rv.std():.3f}')
```

```python
# пример: вычисление третьего квартиля для ранее созданной случайной величины с равномерным распределением

uniform_rv.ppf(0.75)
```
