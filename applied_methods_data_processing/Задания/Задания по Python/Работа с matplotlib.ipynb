# Библиотека matplotlib
Библиотека matplotlib предназначена для визуализации данных - построения двухмерных диаграмм (графиков, гистограмм и .т.п). Для использования этих методов в начале программы (или до первого обращения к методам) необходимо подключить библиотеку. 


Библиотека matplotlib устроена иерархически - она cостоит из множества модулей, наполненных различными классами и функциями, которые иерархически связаны между собой. Для простой работы с matplotlib подключают самый высокоуровневый из модулей - интерфейс matplotlib.pyplot

## Подключение библиотеки и вызов функций
Подключить библиотекy можно несколькими способами. Самый простой - это импорт библиотеки с помощью команды `import`

```python
# основной способ подключения библиотеки (модуля)
import matplotlib.pyplot
```

После этого можно использовать функции из библиотеки. Для их вызова нужно использовать конструкцию      
`<имя библиотеки>.<имя функции>(<параметры>)`    
сперва указывая имя библиотеки, затем, через точку - имя функции. Например, `matplotlib.pyplot.plot(x,y)`

Если название библиотеки длинное, то, чтобы не писать его каждый раз польностью при вызове функции, можно использовать подключение библиотеки c заменой названия библиотеки
import <имя библиотеки> as <псевдоним>

```python
# подключение библиотеки c заменой названия
# де-факто стандарт вызова pyplot
import matplotlib.pyplot as plt
```

## Создание простейшего графика
Пусть имеется список из абсцисс точек

```python
x = [0. , 0.2, 0.4, 0.6, 0.8, 1. , 1.2, 1.4, 1.6, 1.8, 2. , 2.2, 2.4,
       2.6, 2.8, 3. , 3.2, 3.4, 3.6, 3.8, 4. , 4.2, 4.4, 4.6, 4.8, 5. ]
```

и список из ординат точек

```python
y = [ 0.  ,  0.04,  0.16,  0.36,  0.64,  1.  ,  1.44,  1.96,  2.56,
        3.24,  4.  ,  4.84,  5.76,  6.76,  7.84,  9.  , 10.24, 11.56,
       12.96, 14.44, 16.  , 17.64, 19.36, 21.16, 23.04, 25.  ]
```

Для построения графика (ломанная линия, соединяющая соседние точки) используется функция `plot`

```python
plt.plot(x, y)
```

### Оформление графика
Можно задать цвет графика (по названию одного из предустановленных цветов или по коду цвета)

```python
plt.plot(x, y, color = 'green')
```

```python
# для предустановленных цветов достаточно первой буквы названия - r вместо red
# имя аргумента color можно упускать, если цвет указан первым после списков с координатами точек
plt.plot(x, y,'r')  
```

Можно добавить название графика и название осей

```python
plt.plot(x, y)
# название графика
plt.title("График функции $y=x^2$") # знаки $ задают форматирование LaTeX-формулы
# название оси абсцисс
plt.xlabel('$x$')
# название оси ординат
plt.ylabel('$y$')
```

Можно задать стиль рисования линии
* `o` - точки
* `-o` - линия с точками

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, '-o')
```

Можно добавить легенду

```python
# аргумент label задает текст для легенды
plt.plot(x, y, label='$y=x^2$')
# включаем отображение легенды
plt.legend() 
```

Для управления положением легенды используется аргумент `loc` (указывается, в каком из частей внутри координатной плоскости будет расположена легенда)

```python
plt.plot(x, y, label='$y=x^2$')
plt.legend(loc='lower right')
```

Заголовок к легенде добавляется аргументом `title`

```python
plt.plot(x, y, label='$y=x^2$')
plt.legend(title='Функции')
```

## Построение нескольких графиков

При попытке построить графики нескольких функций вы увидите, что они будут построены на одной координатной плоскости

```python
y_1 = [0.00000e+00, 8.00000e-03, 6.40000e-02, 2.16000e-01, 5.12000e-01,
       1.00000e+00, 1.72800e+00, 2.74400e+00, 4.09600e+00, 5.83200e+00,
       8.00000e+00, 1.06480e+01, 1.38240e+01, 1.75760e+01, 2.19520e+01,
       2.70000e+01, 3.27680e+01, 3.93040e+01, 4.66560e+01, 5.48720e+01,
       6.40000e+01, 7.40880e+01, 8.51840e+01, 9.73360e+01, 1.10592e+02,
       1.25000e+02]
```

```python
plt.plot(x, x, label='$y=x$')
plt.plot(x, y, label='$y=x^2$')
plt.plot(x, y_1, label='$y=x^3$')

plt.title('Графики функций') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(title='Функции')
```

Чтобы этого не происходило, нам понадобится сначала задать подложку для области построения графиков, а затем каждому графику указать его место на этой подложке

"Подложка", то есть контейнер для графики `matplotlib`, - это объект `Figure` модуля `pyplot` библиотеки `matplotlib`. За размеры подложки отвечает аргумент `figsize` функции `figure`, который принимает значение кортежа из двух `float` чисел - ширины и высоты подложки

Чтобы указать местоположение каждого графика на подложке используем функцию `subplot` из модуля `pyplot` библиотеки `matplotlib`. У функции `subplot` три аргумента: `subplot(количество строк, количество стоблцов, индекс)`. Подложка разбивается на указанное `количество строк` и `количество стоблцов`, график помещается на место `индекс` (индекс считается слева направо, сверху вниз, начиная с 1)

```python
# создаем подложку с шириной 16 и высотой 5 дюймов
plt.figure(figsize=(16,5))

# указываем местоположение первого графика
# разбиваем подложку на одну строку и три столбца, график помещаем на первую позицию
plt.subplot(1,3,1) 
# cтроим первый график
plt.plot(x, x)
plt.title('График функции $y=x$') 
plt.xlabel('$x$')
plt.ylabel('$y$')

# указываем местоположение второго графика
# повторяем разбиение подложки на одну строку и три столбца, график помещаем на вторую позицию
plt.subplot(1,3,2)
# cтроим второй график
plt.plot(x, y)
plt.title('График функции $y=x^2$') 
plt.xlabel('$x$')
plt.ylabel('$y$')

# указываем местоположение третьего графика
# повторяем разбиение подложки на одну строку и три столбца, график помещаем на вторую позицию
plt.subplot(1,3,3)
# cтроим третий график
plt.plot(x, y_1)
plt.title("График функции $y=x^3$") 
plt.xlabel('$x$')
plt.ylabel('$y$')
```

Рассмотрим пример с четыремя графиками

```python
plt.figure(figsize=(16,10))

# общее название
plt.suptitle('Примеры оформления графиков') 

plt.subplot(2,2,1) 
# аргумент linewidth - задает толщину линии
plt.plot(x, x, color='g', linewidth = 4, label='ключ цвета')
plt.plot(x, y, color='indigo', linewidth = 6, label='название цвета из X11')
plt.plot(x, y_1, color='#8be8da', linewidth = 8, label='RGB')
plt.title('Графики функций') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(title='color')

plt.subplot(2,2,2) 
# аргумент linestyle задает стиль линии, аргумент linewidth - ее толщину
plt.plot(x, x, color='red', linestyle='--', linewidth = 2, label='dashed')
plt.plot(x, y, color='green', linestyle='-.', linewidth = 3, label='dashdot')
plt.plot(x, y_1, color='orange', linestyle=':', linewidth = 4, label='dotted')
plt.title('Графики функций') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(title='linestyle')

plt.subplot(2,2,3) 
# аргумент marker задает вид маркера, аргумент markersize - его размер
plt.plot(x, x, color='red', marker='o', markersize=5, label='точки')
plt.plot(x, y, color='green', marker='X', markersize=9, label='кресты')
plt.plot(x, y_1, color='orange', marker='s', markersize=7, label='квадраты')
plt.title('Графики функций') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(title='marker')

plt.subplot(2,2,4) 
# аргумент markerfacecolor задает цвет маркера, аргумент markeredgecolor - цвет контура маркера

plt.plot(x, x, color='r', marker='o', markersize=9, markerfacecolor='navy', markeredgecolor='skyblue', markeredgewidth = 3, label='точки')
plt.plot(x, y, color='g', marker='X', markersize=12, markerfacecolor='lemonchiffon', markeredgecolor='gold', markeredgewidth = 2, label='кресты')
plt.plot(x, y_1, color='y', marker='s', markersize=9, markerfacecolor='aquamarine', markeredgecolor='teal', markeredgewidth = 3, label='квадраты')
plt.title('Графики функций') 
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(title='marker color')
```

Если аргументы `subplot` меньше 10, то можно использовать синтаксис `subplot(121)` вместо `subplot(1,2,1)`
